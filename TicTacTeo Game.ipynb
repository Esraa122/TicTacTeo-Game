```python
board = ['-', '-', '-',
         '-', '-', '-',
         '-', '-', '-']

playing = True
win = None
c_player = input("Please choose your symbol 'X' or 'O' as a capital letter: ")

if c_player == 'x':
    c_player = 'X'
elif c_player == 'o':
    c_player = 'O'
else:
    c_player = input("Please choose your symbol 'X' or 'O' as capital letters: ")
```

```python
def play():
    display()

    while playing:
        turn(c_player)
        game_over()
        f_player()

    if win == 'X' or win == 'O':
        print(win,"Congratulations!!!\nYou are the winner")
    elif win == None:
        print("Tie")
```

```python
def display():
    print("\n"+board[0] + " | " + board[1] + " | " + board[2])
    print(board[3] + " | " + board[4] + " | " + board[5])
    print(board[6] + " | " + board[7] + " | " + board[8] +"\n")
    print("Positions be like:\n\n1 | 2 | 3\n4 | 5 | 6\n7 | 8 | 9\n")
```

```python
def turn(player):
    print(player, "turn")
    position = input("Choose a position from 1t09: ")

    valid = False
    while not valid:
        while position not in ['1', '2', '3', '4', '5', '6', '7', '8', '9']:
            position = input("Choose a position from 1t09: ")

        position = int(position) - 1

        if board[position] == '-':
            valid = True
        else:
            print("It isn't available")

    board[position] = player

    display()
```

```python
def game_over():
    winner()
    Tie()
```

```python
def winner():
    global win
    
    r_winner = rows()
    col_winner = columns()
    d_winner = diagonals()

    if r_winner:
        win = r_winner
    elif col_winner:
        win = col_winner
    elif d_winner:
        win = d_winner
    else:
        win = None
```

```python
def rows():
    global playing

    r1 = board[0] == board[1] == board[2] != '-'
    r2 = board[3] == board[4] == board[5] != '-'
    r3 = board[6] == board[7] == board[8] != '-'

    if r1 or r2 or r3:
        playing = False

    if r1:
        return board[0]
    elif r2:
        return board[3]
    elif r3:
        return board[6]
    else:
        return None
```

```python
def columns():
    global playing
    
    col1 = board[0] == board[3] == board[6] != '-'
    col2 = board[1] == board[4] == board[7] != '-'
    col3 = board[2] == board[5] == board[8] != '-'

    if col1 or col2 or col3:
        playing = False

    if col1:
        return board[0]
    elif col2:
        return board[1]
    elif col3:
        return board[2]
    else:
        return None
```

```python
def diagonals():
    global playing

    d1 = board[0] == board[4] == board[8] != '-'
    d2 = board[2] == board[4] == board[6] != '-'

    if d1 or d2:
        playing = False

    if d1:
        return board[0]
    elif d2:
        return board[2]
    else:
        return None
```

```python
def Tie():
    global playing

    if '-' not in board:
        playing = False
        return True
    else:
        return False
```

```python
def f_player():
    global c_player

    if c_player == "X":
        c_player = "O"
    elif c_player == "O":
        c_player = "X"
```
